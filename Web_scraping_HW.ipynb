```python
from bs4 import BeautifulSoup as bs
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import requests
import lxml
```

```python
# Setup splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
#set up a url, use splinter to visit it as a test

url = "https://mars.nasa.gov/news/"

browser.visit(url)
```

```python
#scrape the titles and paragraph information off the page at url

response = requests.get(url)

#create the soup object
soup = bs(response.text)


#print it out in pretty form

print(soup.prettify())
```

```python
m_titles = soup.find_all('div', class_ = 'content_title')

for titles in m_titles:
    print(titles.text)


title1 = m_titles[0].text

print(f'The first title found is: {title1}')
```

```python
m_para = soup.find_all('div', class_ = 'rollover_description_inner')

for para in m_para:
    print(para.text)

para1 = m_para[0].text

print(f'The first paragraph found is: {para1}')
```

```python
#find the featured image for JPL featured space create soup object

url2 = "https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars"

browser.visit(url2)

# Click the first link on the page
browser.click_link_by_partial_href('images')
```

```python
#get the image url
#class_ = "BaseButton text-contrast-none w-full mb-5 -primary -compact inline-block"

html = browser.html

ibisque = bs(html, 'html.parser')

print(ibisque.prettify())
```

```python
#find the part of the bisque with the desired url
img_url = ibisque.find_all("img")

print(img_url[2])
```
